```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_parquet("data/processed.parquet", engine="fastparquet")
df["date"] = pd.to_datetime(df["date"])
df = df.set_index("date")

fig, ax1 = plt.subplots(figsize=(12, 5))


ax1.plot(df.index, df["volume"], label="volume")
ax1.set_ylabel("Volume (units)")


ax2 = ax1.twinx()
ax2.plot(df.index, df["price"], color="orange", label="price")
ax2.set_ylabel("Price ($)")


fig.suptitle("Volume vs Price Over Time")
fig.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_parquet("data/processed.parquet", engine="fastparquet")
df["date"] = pd.to_datetime(df["date"])
df = df.set_index("date")

fig, ax1 = plt.subplots(figsize=(12, 5))

# Plot volume
ax1.plot(df.index, df["volume"], label="volume")
ax1.set_ylabel("Volume")

# Start y-axis at 12000
ax1.set_ylim(12000, df["volume"].max())

# Secondary axis for price
ax2 = ax1.twinx()
ax2.plot(df.index, df["price"], color="orange", label="price")
ax2.set_ylabel("Price")

fig.suptitle("Volume vs Price Over Time")
fig.tight_layout()

plt.show()
```
